##### Frequency analysis

```python
# Generating fourier transform of the signal
last_100 = rateE[-1000000:]
n_sample = 100
size = int(len(last_100)/n_sample)
fourrier = []
for i in range(0,n_sample):
    last = last_100[size*i:size*(i+1)]
    fourrier.append(np.fft.rfft(last-np.mean(last)))

fourier = np.array(fourrier)
fourier_freq = np.fft.rfftfreq(size,0.001)
```

```python
plt.plot(fourier_freq,np.mean(np.abs(fourier)**2 ,axis=0))
plt.xlabel("Frequency(Hz)")
plt.ylabel("Amplitude")
```

### Analysis of the response to the pattern

```python
dure_simu = 38000
begining_presentation = 0 # begining of the presentation of the pattern
duree_pattern = 0.1 # duration in second of the pattern
time_step = 0.1 # time step for the pattern presentation
auryn_tstep = 0.0001 # time step of Auryn for the sampling of things like rates in time
nb_neurons = 2048
```

```python
# getting the pattern times from a file in the simulation folder (generated by my code)
all_times = np.genfromtxt(datadir+'/pattern_times', delimiter=' ')
nb_pattern = len(set(all_times[:,1]))
signals = np.zeros((nb_pattern,int(dure_simu//time_step)+1))
signals_times = [[] for i in range(nb_pattern)]
for time,kind in all_times:
    signals_times[int(kind)].append(time)
    signals[int(kind),int((time)//time_step):int((time*auryn_tstep+(duree_pattern))//time_step)]=1
signals_times = np.array(signals_times)
```

```python
number_iter = 10 # number of windows taken to average the activity
time_range = 100 # size of the time window during which the activity is analyzed
nb_signal = 1 # number of different pattern presented to the network
size_window = 0.2 # small window around the pattern presentation  
```

```python
@numba.jit(nopython=True)
def SpikesDistNeurons(spikeTrainN, listsizesN, signal_times, window=0.5,offset=0, nb_neuron = 4096):
    
    spikeTrain = spikeTrainN[:,0]
    neurones = spikeTrainN[:,1]
    
    #listdistN = [[] for _ in range(nb_neuron)] 
    listdistN = np.empty((nb_neuron,10000,2))

    r = 0

    for event in signal_times:
        for spike_i in range(r, len(spikeTrain)):
            if spikeTrain[spike_i] > (event+window/2)+offset:
                break
            if spikeTrain[spike_i] < (event-window/2)+offset:
                r += 1
            else:
                neuron = int(neurones[spike_i])
                mtuple = spikeTrain[spike_i]-event,spikeTrain[spike_i]
                listdistN[neuron,listsizesN[neuron]] = mtuple
                listsizesN[neuron]+=1


    return listdistN,listsizesN

def PSTH(spikes,wch_pat,nb_neurons,time_range,signals_times,size_window,duree_pattern):
    spikes_p= spikes
    dis_p,tailles_p = SpikesDistNeurons(spikes_p,np.zeros((nb_neurons,),dtype=np.int),np.array(signals_times[wch_pat]),window = size_window,offset=duree_pattern/2,nb_neuron=nb_neurons)

    all_dis_p = []
    for d in range(len(dis_p)):
        all_dis_p.append(dis_p[d,:tailles_p[d]])
    
    data_p=all_dis_p
    dist_p = []
    tms_p = []
    for i_p in  data_p:
        for j_p in i_p:    
            dist_p.append(j_p[0])
            tms_p.append(j_p[1])
    
    return data_p,spikes_p,dist_p,tms_p

def all_PSTH(number_iter,time_range,nb_signal,size_window,sfo,dure_simu,signals_times,nb_neurons,duree_pattern,starting_time = 0):
    spikes_in_time = dict()
    dist_in_time = dict()
    times_in_time = dict()
    data_in_time = dict()
    dure_simu = dure_simu-starting_time

    for wch_pat in range(nb_signal):
        spikes_in_time[wch_pat] = dict()
        dist_in_time[wch_pat] = dict()
        times_in_time[wch_pat] = dict()
        data_in_time[wch_pat] = dict()
        times = []
        tms = [T for T in range(starting_time+int(dure_simu/number_iter)-time_range,starting_time+dure_simu,int(dure_simu/number_iter))]
        print("extracting_spikes...") 
        all_spikes = []
        for T in tms:
            all_spikes.append(np.array(sfo.get_spikes(t_start=T,t_stop=T+time_range)) )

        print("computing...")
        for letem in range(len(tms)):
            print(letem)
            data_in_time[wch_pat][tms[letem]],spikes_in_time[wch_pat][tms[letem]],dist_in_time[wch_pat][tms[letem]],times_in_time[wch_pat][tms[letem]]=PSTH(all_spikes[letem],wch_pat,nb_neurons,time_range,signals_times,size_window,duree_pattern)
        
        times=list(data_in_time[wch_pat].keys())
    return spikes_in_time,dist_in_time,times_in_time,data_in_time,times

all_spikes_in_time,all_dist_in_time,all_times_in_time,all_data_in_time,times = all_PSTH(number_iter,time_range,nb_signal,size_window,sfo,dure_simu,signals_times,nb_neurons,duree_pattern)

```

```python
bin_size = 1e-3
which_signal=0
spikes_in_time = all_spikes_in_time[which_signal]
dist_in_time = all_dist_in_time[which_signal]
times_in_time = all_times_in_time[which_signal]
data_in_time = all_data_in_time[which_signal]
```

```python
H, bins = np.histogram(dist_in_time[times[0]],bins=150);
plt.bar(bins[:-1],H,width=(1/(len(bins)*1.9)))
plt.show()
```

```python
Htot, bins = np.histogram(dist_in_time[times[-1]],bins=250);

moyenne_plus = np.mean(Htot) + 1*np.std(Htot)
delimiters = []
moyenne = np.mean(Htot)

for bn in range(len(Htot)):
    
    value = Htot[bn]

    if value > moyenne_plus:

        left = bn
        while not left<0 and (Htot[left]<= Htot[left+1] or Htot[left]>moyenne_plus) :
            left-=1
            
        right = bn
        while not right>=len(Htot) and ( Htot[right]<= Htot[right-1] or Htot[right]>moyenne_plus):
            right+=1
    
        delimiters.append((left,right))
            
```

```python
plt.bar(bins[:-1],Htot,width=(1/(len(bins)*2)))
plt.plot([np.min(bins),np.max(bins)],[moyenne_plus,moyenne_plus])
```

```python
pics_bins = np.array(list(set(delimiters)))
pics_bins_changed = np.array([pics_bins[:,0],pics_bins[:,1]]).T
times_pics = bins[:-1][pics_bins_changed]
distancesN = data_in_time[times[-1]]
scores_pics = []
for tms in times_pics:
    scores_pics.append(np.array([ np.sum(np.logical_and(np.array(x)[:,0]<tms[1],np.array(x)[:,0]>tms[0]))/len(x) if len(x)>0 else 0 for x in distancesN]))

percentage_pic = []
for tms in times_pics:
    percentage_pic.append((tms[1]-tms[0])/size_window)

in_pics = []
for i in range(len(percentage_pic)):
    perc = percentage_pic[i]
    pic_ones = []
    for j in range(len(scores_pics[i])):
        score = scores_pics[i][j]
        if score > perc*2: # CHANGED
            pic_ones.append(j)
    in_pics.append(pic_ones)
```

```python
def take_spikes_pics(distancesN):
    spikes_in_pics = []
    for pic_ones in in_pics:
        spikes_in_pics.append(np.array([]))
        for one in pic_ones :
            train = np.array(distancesN[one])[:,0]
            spikes_in_pics[-1]=np.append(spikes_in_pics[-1],train)
    return spikes_in_pics
spikes_in_pics=take_spikes_pics(distancesN)
```
